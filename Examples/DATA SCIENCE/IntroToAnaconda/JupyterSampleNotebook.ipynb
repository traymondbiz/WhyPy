```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('anscombe.csv')
data.head(15)
```

```python
fig, axs = plt.subplots(1, 2, sharey=False)
fig2, axs2 = plt.subplots(1, 2, sharey=False)
data.plot(kind='scatter', x='X1', y='Y1', ax=axs[0])
data.plot(kind='scatter', x='X2', y='Y2', ax=axs[1])
data.plot(kind='scatter', x='X3', y='Y3', ax=axs2[0])
data.plot(kind='scatter', x='X4', y='Y4', ax=axs2[1])
```

```python
seadata = sns.load_dataset("anscombe")

sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=seadata,
           col_wrap=2, ci=None, palette="dark")
```
